```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier #랜덤포레스트라는 분류기를 사용하기 위해 import
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
import multiprocessing
from sklearn.model_selection import GridSearchCV
import pandas as pd
import seaborn as sns
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import matplotlib.pyplot as plt
```

```python
"""
data의 feature을 통해서 유방암이 악성인지 음성인지 판단하는 분류 문제 이다 

크기는 569개 정도 
"""

cancer = load_breast_cancer()
cancer_data = cancer.data
cancer_label = cancer.target
print(cancer.keys())
print(cancer["DESCR"])

print(cancer_data.shape)
```

```python
# 데이터 시각화 하기 
cancer_df = pd.DataFrame(cancer.data,columns=cancer.feature_names)
cancer_result = pd.Series(cancer.target,dtype="category")
cancer_result = cancer_result.cat.rename_categories(cancer.target_names) # 카테고리로 target 이름 재 지정하기 
cancer_df["Target"] = cancer_result
cancer_df.head()
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=9)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
# len은 배열의 길이를 출력
```

```python
model = make_pipeline(MinMaxScaler(),LogisticRegression())
model.fit(X_train,y_train)
```

```python
print("학습 데이터 점수: {} ".format(model.score(X_train,y_train)))
print("평가 데이터 점수: {} ".format(model.score(X_test,y_test)))
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=9)
model = make_pipeline(StandardScaler(),LogisticRegression())
model.fit(X_train,y_train)
```

```python
print("학습 데이터 점수: {} ".format(model.score(X_train,y_train)))
print("평가 데이터 점수: {} ".format(model.score(X_test,y_test)))
```

### 정규화 보다 표준화 했을때 성능이 더 좋았다 


### 성능 평가 
- 오차 행렬 
- 정확도 및 f1 score 
- cross_val_score 
- classification_report(y_test, y_pred)


```python
# 표준화 + 로지스틱 회귀 오차 행렬 및 성능 평가 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score,recall_score

predict = model.predict(X_test)
print("정확도 {} ".format(accuracy_score(y_test,predict)))
print("정밀도 {}".format(precision_score(y_test,predict)))
print("재현율(recall) {}".format(recall_score(y_test,predict)))

# 높은 정밀도 정확도 가진다 recall도 1이니까 best value이다 
```

```python
# 오차 행렬 

confmat = confusion_matrix(y_true=y_test,y_pred=predict)

print(confmat)

"""
TN FP
FN TP
개수 행렬이다 

TN과 TP가 높은것을 알수있다 
"""
```

```python
# precision과 recall이 어느 한쪽에 치우치지 않았는지 확인하는 평가 지표 F1 score
from sklearn.metrics import f1_score
f1=f1_score(y_test,predict)
f1  
```

```python
# cross val score 했을때 성능 평가 

from sklearn.model_selection import cross_val_score,cross_validate
import numpy as np

X,y=load_breast_cancer(return_X_y=True)  #표준화 포함한 model이다 

scores = cross_val_score(model,X,y,cv=10)
print("교차검증 정확도: {}".format(scores))
print("교차검증 정확도 평균: {}".format(np.mean(scores)))
```

```python
# stratified k fold 
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=10,shuffle=True)
scores_skf = cross_val_score(model,X,y,cv=skf)
print("StratifiedKFold 정확도: {}".format(scores_skf))
print("StratifiedKFold 정확도 평균: {}".format(np.mean(scores_skf)))
```

### StratifiedKFold 보다 cross_val_score가 더 좋게 나왔다

```python
# 최적의 파라미터 구하기 GrideSearchCV
X,y=load_breast_cancer(return_X_y=True)
param_grid = {"penalty":["l1", "l2", "elasticnet", "none"],
             "C":[0.1,0.2,1.0,2.0,5.0,10.0,100.0],
              "max_iter":[100,3000,5,500,1000],
             "solver":["newton-cg", "lbfgs", "liblinear", "sag", "saga"]}

gs = GridSearchCV(estimator=LogisticRegression(),param_grid=param_grid,cv=10,
                  n_jobs=multiprocessing.cpu_count())

gs
```

```python
result = gs.fit(X,y)

pd.DataFrame(result.cv_results_)
```

```python
print(gs.best_estimator_)
print("최적의 점수 {}".format(gs.best_score_))
print("최적의 파라미터 {}".format(gs.best_params_))
```

## GrideSearchCV에 다른 파라미터 값을 넣어주어서 성능이 낮게 나온 것 같다 

## SGDClassifier() 보기 

```python
from sklearn.linear_model import SGDClassifier
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=9)
```

```python
#Grid search 해보기 SGDClassifier() 파라미터 설정하기 
param_grid = {"loss":["log","hinge","modified_huber"],
              "penalty":["l1", "l2", "elasticnet"],
              "max_iter":[200,3000,5,500,1000],
              "shuffle":[True,False],
              "alpha":[0.0001,0.001,0.1,0.00001]
             }

gs = GridSearchCV(estimator=SGDClassifier(),param_grid=param_grid,cv=10,
                  n_jobs=multiprocessing.cpu_count())

```

```python
result = gs.fit(X,y)
```

```python
print("최적 점수 {}".format(result.best_score_))
print("최적 파라미터 {}".format(result.best_params_))
```

```python
result.best_estimator_
```

# 유방암 데이터 셋 회고 

가슴 둘레  가슴 반지름등의 데이터로 악성인지 음성인지 분류하는 문제 였다 

정규화 했을때 보다 표준화 했을때가 성능이 더 좋았다 이상치가 많이 있었던것 같다 

이진 분류 문제여서 LogisticRegreesion 했을때 성능이 좋게 나온것 같다 
정확도 ,정밀도,f1 score ,recall score 모두 좋은 점수가 나왔다 

표준화와 logisticregression했을때 좋은 성능 나왔다 cross val score를 했을때도 statifiedKFold 방식이 사용되어서 좋은 
성능이 나왔다 statifiedKFold는 분류 문제에서 클래스 비율까지 고려 하기 때문에 k fold 보다 더 좋은 검증 방식이다 

일반 말고 shuffle 기능까지 추가한 stratifiedKFold를 cv로 직접 설정했을때 비슷하게 나왔지만 평균만 보면 그냥 cross val score 한게 더 좋은 성능이 나왔다

grid searchCV로 최적의 파라미터를 찾을것이다. C는 규제 강도이다 값이 작을수록 작아진다 
C 가 높으면 train data에 대한 성능이 높은것이고 낮으면 신뢰성이 낮다고 볼수있다. max_iter는 최대 반복하는 횟수이다 
최적의 파라미터를 보면 C의 값이 설정 한 값들 중에서 가장 높은 값을 가진다 즉 train data에 대한 신뢰성이 높은 편이라고 할수있다
 최적의 파라미터를 찾아도 성능이 일반으로 돌렸을때보다 낮은 성능이 나왔다 scoring을 accuacy로 해서 그런것으로 예상된다 
scoring을 f1으로 바꾸니까 성능이 좋아졌다 그래도 여전히 일반으로 했을때보다는 낮은 성능을 보였다 

시간이 많이 걸린다는게 단점이다 
SGDClassifier() 학습 파라미터에 대한 손실함수의 기울기를 구해 기울기가 최소화가 되는 방향으로 학습한다  기본 loss function이  hinge로 svm linear이다  
최적의 파라미터를 찾기 위해 grideSearchCV 사용했다 이상치가 많아서 그런지 loss function은 modified huber가 나왔는데 
이는 이상치를 확률 추정치로 대체하는 loss function이다 하지만 성능은 그렇게 좋지 못했다 
오히려 loss function을 log 즉 logistic function으로 설정하고 데이터를 섞어주는게 더 좋은 성능을 보였다 
평가 데이터에 대해서 99%의 성능을 보여준다 이 방법이 가장 좋은 성능이었다 

SVM이나 decision tree 같은경우 아직 이해도가 높지 않아서 lms에 내용 그대로 출력 하고 성능 측정 하였다
