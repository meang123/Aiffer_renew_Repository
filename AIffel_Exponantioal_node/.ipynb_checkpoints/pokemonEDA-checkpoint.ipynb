## 포켓몬 이름 모두 알파벳 문자로 전처리 하기 


포켓몬 이름 타입 확인하기   
1. 한 단어면 ex. Venusaur
2. 두 단어이고, 앞 단어는 두 개의 대문자를 가지며 대문자를 기준으로 두 부분으로 나뉘는 경우 ex. VenusaurMega Venusaur
3. 이름은 두 단어이고, 맨 뒤에 X, Y로 성별을 표시하는 경우 ex. CharizardMega Charizard X
4. 알파벳이 아닌 문자를 포함하는 경우 ex. Zygarde50% Forme


```python
# NAME nospace 컬럼 추가 해서 띄어쓰기 없애준다 
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))
pokemon.tail()
```

```python
# 문자열로만 이루어져있는지 확인하기 
pokemon["name_isalpha"] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())
pokemon.head()
```

```python
# 알파벳으로만 구성 되어이있지 않은 포켓몬 이름 확인하기 
print(pokemon[pokemon["name_isalpha"] == False].shape) # 9개가 존재 한다 
pokemon[pokemon["name_isalpha"] == False] 
```

```python
# 문자열 이름으로 하나씩 변경 하기 
pokemon = pokemon.replace(to_replace="Nidoran♀", value="Nidoran X")
pokemon = pokemon.replace(to_replace="Nidoran♂", value="Nidoran Y")
pokemon = pokemon.replace(to_replace="Farfetch'd", value="Farfetchd")
pokemon = pokemon.replace(to_replace="Mr. Mime", value="Mr Mime")
pokemon = pokemon.replace(to_replace="Porygon2", value="Porygon")
pokemon = pokemon.replace(to_replace="Ho-oh", value="Ho Oh")
pokemon = pokemon.replace(to_replace="Mime Jr.", value="Mime Jr")
pokemon = pokemon.replace(to_replace="Porygon-Z", value="Porygon Z")
pokemon = pokemon.replace(to_replace="Zygarde50% Forme", value="Zygarde Forme")

pokemon.loc[[34, 37, 90, 131, 252, 270, 487, 525, 794]]
```

## 정규화로 token 분리 하기 

```python

import re 

def tokenize(name):
    name_split = name.split(" ")
    
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
        
    return np.array(tokens)
```

```python
# tokenize 함수의 간단한 사용 예제 
name = "CharizardMega Charizard X"
tokenize(name)
```

```python
# 전설 포켓몬 이름 token 하기 
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
# 전설 포켓몬 이름들중 가장 많이 사용되는 이름 형식 
from collections import Counter
most_common = Counter(token_set).most_common(10) # 가장 많이 사용되는 이름 형식 
most_common
```

```python
# 이름 포함 하는지 확인하기 
for token, _ in most_common:
    # pokemon[token] = ... 형식으로 사용하면 뒤에서 warning이 발생합니다
    pokemon[f"{token}"] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

## Type 속성 one hot encoding 하기 

integer encoding과 one hot encoding이 대표적인 범주형 데이터 처리 방식이다 

integer encoding 같은 경우 대소 관계가 있기 때문에 예를 들어 만족 매우 만족 불만 매우 불만 등의 관계가 있는 범주형에는 적합하지만 그외에는 적합하지 않다 

<br>
</br>

one hot encoding은 모든 범주를 새로운 컬럼으로 만들고 해당 범주에 1을 활성화 하는 방법 
컬럼이 많아진다는 단점이 있다 


```python
types = ['Poison', 'Flying', 'Psychic', 'Ghost', 'Steel', 'Fairy', 'Fighting', 'Normal', 'Electric', 'Bug', 'Fire', 'Water', 'Ground', 'Rock', 'Dragon', 'Ice', 'Dark', 'Grass']
```

```python
# one hot encoding of type1 type2 
for t in types:
    pokemon[t] = (pokemon["Type 1"] == t) | (pokemon["Type 2"] == t)
    
```

## 전설과 일반 포켓몬 분류 하는 머신러닝 데이터 전처리 하기 


```python
 #원본 데이터로 처리 한다 
original_data.columns
```

```python
# featur와 target 나누기 
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
target = 'Legendary'
```

```python
# X,y = load_boston(returnX,y=True) 와 같은 작업 
X = original_data[features]
print(X.shape)
X.head()
```

```python
y = original_data[target]
print(y.shape)
y.head()
```

```python
# 데이터 분리 하기 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## DecisionTreeClassifier 분류 모델 사용하기 

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=25)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 의사결정 나무 모델 점수 
print("학습데이터 점수 : {} ".format(model.score(X_train,y_train)))
print("평가데이터 점수 : {} ".format(model.score(X_test,y_test)))
```

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
from plotly import tools
import plotly.graph_objs as go
```

```python
# Function that plots the stat of a pkmn
def hex_stat(data, xx):
    x = data[data['Name'] == xx]
    data = [go.Scatterpolar(
        r = [x['HP'].values[0],x['Attack'].values[0],x['Defense'].values[0],x['Sp. Atk'].values[0],x['Sp. Def'].values[0],x['Speed'].values[0],x["HP"].values[0]],
        theta = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','HP'],fill = 'toself')]

    layout = go.Layout(polar = dict(
        radialaxis = dict(
            visible = True,
            range = [0, 250]
        )
    ),showlegend = False,
                       title = "Stats of {}".format(x.Name.values[0]))
    
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename = "Single Pokemon stats")
```

```python
# Creating a method to compare 2 pokemon
def compare2pokemon(data, xx,yy):
    x = data[data['Name'] == xx]
    y = data[data['Name'] == yy]

    trace0 = go.Scatterpolar(
      r = [x['HP'].values[0],x['Attack'].values[0],x['Defense'].values[0],x['Sp. Atk'].values[0],x['Sp. Def'].values[0],x['Speed'].values[0],x["HP"].values[0]],
      theta = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','HP'],
      fill = 'toself',
      name = x.Name.values[0]
    )

    trace1 = go.Scatterpolar(
      r = [y['HP'].values[0],y['Attack'].values[0],y['Defense'].values[0],y['Sp. Atk'].values[0],y['Sp. Def'].values[0],y['Speed'].values[0],y["HP"].values[0]],
      theta = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','HP'],
      fill = 'toself',
      name = y.Name.values[0]
    )

    data = [trace0, trace1]

    layout = go.Layout(
      polar = dict(
        radialaxis = dict(
          visible = True,
          range = [0, 200]
        )
      ),
      showlegend = True,
      title = "{} vs {}".format(x.Name.values[0],y.Name.values[0])
    )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename = "Two Pokemon stats")
```

```python
# 원픽들의 ststus 시각화 하기 
# Psyduck , Mew,Eevee,Bulbasaur, Pikachu

# 수연님의 포켓몬 
hex_stat(pokemon, 'Psyduck')
```

```python
# 주완님 포켓몬의 능력치 
hex_stat(pokemon, 'Bulbasaur')
```

```python
# 상준님의 포켓몬 능력치 
hex_stat(pokemon, 'Mew')
```

```python
# 성주의 포켓몬 능력치 
hex_stat(pokemon, 'Pikachu')
```

```python
# 광석님 이브이!
hex_stat(pokemon, 'Eevee')
```

```python
# 성주와 상준님의 포켓몬 승부는?
compare2pokemon(pokemon,"Pikachu","Mew")
```
